```python
import speech_recognition as sr
import os
import librosa
import webbrowser
import openai
import datetime
import random
import pyttsx3
import wikipedia
import pywhatkit as kit
import sys
from googlesearch import search


engine = pyttsx3. init()
engine. setProperty('rate', 150)
voices = engine. getProperty('voices')
engine

    
def say(text):
    engine.say(text)
    engine.runAndWait()
    os.system(f"say {text}")

    
def wish():
    hour=int(datetime.datetime.now().hour)
    
    if hour>=6 and hour<=12:
        print("Good Morning")
        say("Good Morning")
    elif hour>12 and hour<18:
        print("Good Afternoon")
        say("Good afternoon")
    else :
        print("Good Evening")
        say("Good evening")
        
def takeCommand():
    
    r=sr.Recognizer()
    with sr.Microphone() as source:
        r.pause_threshold =0.8
        audio=r.listen(source)
        try:
            print("recognising")
            query=r.recognize_google(audio, language="en-in")
            print(f"User said: {query}")
            say(query)
            return query
        
        except Exception as e:
            print("I am unable to hear you. Please say again.")
            say("I am unable to hear you. Please say again.")
            return "I am unable to hear you. Please say again."
      

if  __name__ == '__main__':
    wish()
    print('at your service Sir')
    say(' at your service Sir')
    say("Hello I am DAS. I can do many things for you.I What can I do for you?")
    while True:
        
        print("Listening...")
        query=takeCommand()
        sites=[["YouTube", "https://www.youtube.com"],["Wikipedia", "https://www.wikipedia.com"],["Google","https://www.google.com"],["Instagram","https://www.instagram.com"],[f"{query}",f"https://www.{query}.com"],]
        for site in sites:
            if f"Open {site[0]}".lower() in query.lower():
                say(f"Opening {site[0]}")
                webbrowser.open(site[1])
        
        if "the time".lower() in query.lower():
            strfTime= datetime.datetime.now().strftime("%H:%M:%S")
            say(f"The time is {strfTime}")
            print(f"the time is {strfTime}")
       
        elif "open vsd".lower() in query.lower():
                npath="C:\\Users\\KIIT\\AppData\\Local\\Programs\\Microsoft VS Code"
                os.startfile(npath)  
                
        elif "what is".lower() in query.lower():
            
            print("Let me think....\n")
            say("Let me think\n")
            query=query.replace("what is", "")
            reply=wikipedia.summary(query,sentences=2)
            print("So I found that")
            say("So I found that")
            print(f"{reply}")
            say(reply)
            
        elif "search in google".lower() in query.lower():
            say("What should I search")
            sg=takeCommand()
            print(f"searching {sg}")
            say(f"searching {sg}")
            webbrowser.open(f"{sg}")
         
        elif "send message".lower() in query.lower():
            kit.sendwhatmsg("+918210509460","hi",1,7)
          
        elif "open a video".lower() in query.lower():
            say("what should I play")
            ov=takeCommand()
            print(f"playing {ov}")
            say(f"playing {ov}")
            kit.playonyt(f"{ov}")
         
        elif "Take rest".lower() in query.lower():
            say("Had a great time interrracting with you")
            say("By by. Have a great day")
            say("Switching off")
            print("----------POWER OFF----------")
            sys.exit()
            
        say("Any more service required Sir?")      
        print("Any more service required Sir?")    

      
        
          
            
            
           
```
